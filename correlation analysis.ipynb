```python
import pyarrow.parquet as pq
import pandas as pd
```

```python
prom_file_path = "/home/wenjun/r11n5_continuous_data.parquet"
data = pd.read_parquet(prom_file_path)
```

```python
data = data.drop(columns=['node'])
```

```python
# Pearson correlation
pearson_corr = data.corr(method='pearson')
pearson_corr_power = pearson_corr['surfsara_power_usage']
pearson_corr_load = pearson_corr['node_load1']

# Spearman correlation
spearman_corr = data.corr(method='spearman')
spearman_corr_power = spearman_corr['surfsara_power_usage']
spearman_corr_load = spearman_corr['node_load1']
```

```python

import matplotlib.pyplot as plt

# Visualization function
def plot_correlation(correlation_data, title):
    plt.figure(figsize=(12, 6))
    correlation_data.sort_values().plot(kind='bar')
    plt.title(title)
    plt.ylabel('Correlation Coefficient')
    plt.show()

# Visualize Pearson correlation for 'surfsara_power_usage'
plot_correlation(pearson_corr_power, "Pearson Correlation with surfsara_power_usage")

# Visualize Spearman correlation for 'surfsara_power_usage'
plot_correlation(spearman_corr_power, "Spearman Correlation with surfsara_power_usage")

# Visualize Pearson correlation for 'node_load1'
plot_correlation(pearson_corr_load, "Pearson Correlation with node_load1")

# Visualize Spearman correlation for 'node_load1'
plot_correlation(spearman_corr_load, "Spearman Correlation with node_load1")

```
