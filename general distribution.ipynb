```python
import pyarrow.parquet as pq
import pandas as pd
```

```python
prom_file_path = "/mnt/sdc/xiaoyu/prom_extr_att.parquet"
data = pd.read_parquet(prom_file_path)
```

```python
# Descriptive Statistics
print("Descriptive Statistics for surfsara_power_usage:")
print(data['surfsara_power_usage'].describe())

print("\nDescriptive Statistics for node_load1:")
print(data['node_load1'].describe())
```

```python
# Quantiles
quantiles = [0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99]

print("Quantiles for surfsara_power_usage:")
print(data['surfsara_power_usage'].quantile(quantiles))

print("\nQuantiles for node_load1:")
print(data['node_load1'].quantile(quantiles))
```

```python
print("Missing values in surfsara_power_usage:", data['surfsara_power_usage'].isna().sum())
print("Missing values in node_load1:", data['node_load1'].isna().sum())
```

```python
import matplotlib.pyplot as plt

# Histogram for surfsara_power_usage
plt.figure(figsize=(10,5))
data['surfsara_power_usage'].hist(bins=50)
plt.xlabel("Power Usage")
plt.ylabel("Frequency")
plt.grid(False)
plt.show()

# Histogram for node_load1
plt.figure(figsize=(10,5))
data['node_load1'].hist(bins=50)
plt.xlabel("Node Load")
plt.ylabel("Frequency")
plt.grid(False)
plt.show()
```
