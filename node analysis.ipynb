```python
import pyarrow.parquet as pq
import pandas as pd
```

```python
prom_file_path = "/mnt/sdc/xiaoyu/prom_extr_att.parquet"
data = pd.read_parquet(prom_file_path)
```

```python
grouped_means = data.groupby('node').mean()
```

```python
top_10_power = grouped_means.nlargest(10, 'surfsara_power_usage')['surfsara_power_usage']
least_10_power = grouped_means.nsmallest(10, 'surfsara_power_usage')['surfsara_power_usage']

top_10_load = grouped_means.nlargest(10, 'node_load1')['node_load1']
least_10_load = grouped_means.nsmallest(10, 'node_load1')['node_load1']
```

```python
import matplotlib.pyplot as plt

def plot_means(data, title):
    plt.figure(figsize=(12, 6))
    data.plot(kind='bar')
    plt.title(title)
    plt.ylabel('Mean Value')
    plt.show()

plot_means(top_10_power, "Top 10 Nodes by Mean Power Usage")
plot_means(least_10_power, "Least 10 Nodes by Mean Power Usage")
plot_means(top_10_load, "Top 10 Nodes by Mean Load")
plot_means(least_10_load, "Least 10 Nodes by Mean Load")
```
